```python
tp = HTMLTableParser()
```

```python
from time import sleep
for block_no,loc in zip(station_info.number[3:],station_info.location[3:]):
    for view_num in view_info.view_number:
        page = requests.get(f'{baseurl}block_no={block_no}&view={view_num}')
        soup = BeautifulSoup(page.content,'html.parser')
        table = soup.find_all('table')[1] 
        tp.parse_html_table(table).to_csv(f'{loc}_{block_no}_{view_num}.csv')
        sleep(1)
```
